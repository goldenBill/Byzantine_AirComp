```python
import pickle
import numpy as np
import matplotlib.pyplot as plt

marktitle = ""
CACHE_DIR = './MNIST_Air_weight_MLP_Final/' + 'mnist_K50_B5_DataParallel' + '_'
with open(CACHE_DIR + 'SGD_classflip_gm2'+''+marktitle, 'rb') as f:
    obj = pickle.load(f)
    acc_1 = obj['valAccPath']
    loss_1 = obj['valLossPath']

marktitle = ""
CACHE_DIR = './MNIST_Air_weight_MLP_Final/' + 'mnist_K50_B10_DataParallel' + '_'
with open(CACHE_DIR + 'SGD_classflip_gm2'+''+marktitle, 'rb') as f:
    obj = pickle.load(f)
    acc_2 = obj['valAccPath']
    loss_2 = obj['valLossPath'] 
print('data_load finished!')

marktitle = ""
CACHE_DIR = './MNIST_Air_weight_MLP_Final/' + 'mnist_K50_B5_DataParallel' + '_'
with open(CACHE_DIR + 'SGD_classflip_gm_0.01'+''+marktitle, 'rb') as f:
    obj = pickle.load(f)
    acc_1_n = obj['valAccPath']
    loss_1_n = obj['valLossPath']

marktitle = ""
CACHE_DIR = './MNIST_Air_weight_MLP_Final/' + 'mnist_K50_B10_DataParallel' + '_'
with open(CACHE_DIR + 'SGD_classflip_gm_0.01'+''+marktitle, 'rb') as f:
    obj = pickle.load(f)
    acc_2_n = obj['valAccPath']
    loss_2_n = obj['valLossPath']
print('data_load finished!')

marktitle = ""
CACHE_DIR = './MNIST_Air_weight_MLP_Final/' + 'mnist_K50_B5_DataParallel' + '_'
with open(CACHE_DIR + 'SGD_weightflip_gm2'+''+marktitle, 'rb') as f:
    obj = pickle.load(f)
    acc_3 = obj['valAccPath']
    loss_3 = obj['valLossPath']

marktitle = ""
CACHE_DIR = './MNIST_Air_weight_MLP_Final/' + 'mnist_K50_B10_DataParallel' + '_'
with open(CACHE_DIR + 'SGD_weightflip_gm2'+''+marktitle, 'rb') as f:
    obj = pickle.load(f)
    acc_4 = obj['valAccPath']
    loss_4 = obj['valLossPath']
print('data_load finished!')


marktitle = ""
CACHE_DIR = './MNIST_Air_weight_MLP_Final/' + 'mnist_K50_B5_DataParallel' + '_'
with open(CACHE_DIR + 'SGD_weightflip_gm_0.01'+''+marktitle, 'rb') as f:
    obj = pickle.load(f)
    acc_3_n = obj['valAccPath']
    loss_3_n = obj['valLossPath']

marktitle = ""
CACHE_DIR = './MNIST_Air_weight_MLP_Final/' + 'mnist_K50_B10_DataParallel' + '_'
with open(CACHE_DIR + 'SGD_weightflip_gm_0.01'+''+marktitle, 'rb') as f:
    obj = pickle.load(f)
    acc_4_n = obj['valAccPath']
    loss_4_n = obj['valLossPath']
print('data_load finished!')
```

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(16*4/4, 9/4 +1), constrained_layout=True)

# Labels to use in the legend for each line
line_labels = ["B = 5 AirComp", "B = 5 Ideal", "B = 10 AirComp", "B = 10 Ideal"]

import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
# Create the sub-plots, assigning a different color for each line.
# Also store the line objects created

color1 = '#990033'
color2 = '#024C7C'
color3 = '#8CA85A'
# color1 = "mediumblue"
# color2 = "darkorange"
color1 = "mediumblue"
color2 = "#D95319"

l1_loss_n = ax1.plot(np.arange(0,101)*10, loss_1_n[0:101], color=color1 )[0]
l1_loss = ax1.plot(np.arange(0,101)*10, loss_1[0:101], color=color1, linestyle='--')[0]
l2_loss_n = ax1.plot(np.arange(0,101)*10, loss_2_n[0:101], color=color2)[0]
l2_loss = ax1.plot(np.arange(0,101)*10, loss_2[0:101], color=color2, linestyle='--')[0]
ax1.set_xlabel('Iteration', fontsize=12)
ax1.set_ylabel('Test Loss', fontsize=12)
ax1.grid(True)
ax1.set_yticks([0.25, 0.30, 0.35, 0.40,])
ax1.set_ylim([0.2605, 0.425])
ax1.set_title('Class Flip Attack')


l1_acc_n = ax2.plot(np.arange(0,101)*10, acc_1_n[0:101], color=color1)[0]
l1_acc = ax2.plot(np.arange(0,101)*10, acc_1[0:101], color=color1, linestyle='--')[0]
l2_acc_n = ax2.plot(np.arange(0,101)*10, acc_2_n[0:101], color=color2)[0]
l2_acc = ax2.plot(np.arange(0,101)*10, acc_2[0:101], color=color2, linestyle='--')[0]
ax2.set_xlabel('Iteration', fontsize=12)
ax2.set_ylabel('Test Accuracy', fontsize=12)
ax2.grid(True)
ax2.set_yticks([0.88, 0.89, 0.90, 0.91, 0.92])
ax2.set_ylim([0.89, 0.925])
ax2.set_title('Class Flip Attack')

l3_loss_n = ax3.plot(np.arange(0,101)*10, loss_3_n[0:101], color=color1)[0]
l3_loss = ax3.plot(np.arange(0,101)*10, loss_3[0:101], color=color1, linestyle='--')[0]
l4_loss_n = ax3.plot(np.arange(0,101)*10, loss_4_n[0:101], color=color2)[0]
l4_loss = ax3.plot(np.arange(0,101)*10, loss_4[0:101], color=color2, linestyle='--')[0]
ax3.set_xlabel('Iteration', fontsize=12)
ax3.set_ylabel('Test Loss', fontsize=12)
ax3.grid(True)
ax3.set_yticks([0.25, 0.30, 0.35, 0.40,])
ax3.set_ylim([0.2605, 0.425])
ax3.set_title('Weight Flip Attack')

l3_acc_n = ax4.plot(np.arange(0,101)*10, acc_3_n[0:101], color=color1)[0]
l3_acc = ax4.plot(np.arange(0,101)*10, acc_3[0:101], color=color1, linestyle='--')[0]
l4_acc_n = ax4.plot(np.arange(0,101)*10, acc_4_n[0:101], color=color2)[0]
l4_acc = ax4.plot(np.arange(0,101)*10, acc_4[0:101], color=color2, linestyle='--')[0]
ax4.set_xlabel('Iteration', fontsize=12)
ax4.set_ylabel('Test Accuracy', fontsize=12)
ax4.grid(True)
ax4.set_yticks([0.88, 0.89, 0.90, 0.91, 0.92])
ax4.set_ylim([0.89, 0.925])
ax4.set_title('Weight Flip Attack')

# Create the legend
fig.legend([l1_loss_n, l1_loss, l2_loss_n, l2_loss],     # The line objects
           labels=line_labels,   # The labels for each line
           loc="lower center",   # Position of legend
           borderaxespad=0.1,    # Small spacing around legend box
           ncol = 4,
           )

# Adjust the scaling factor to fit your legend text completely outside the plot
# (smaller value results in more space being made for the legend)

plt.subplots_adjust(left=None, bottom=0.22, right=None, top=0.84,
                wspace=0.28, hspace=None)
# plt.savefig("./imag/total_acc_Loss.eps", bbox_inches='tight')
plt.show()
```
